# **1. Extraction: "I want to die"**. 

```python
import twint
# Set up TWINT config
c = twint.Config()
c.Search = "I want to die"
# Custom output format
c.Limit = 3000
c.Pandas = True
twint.run.Search(c)
```

```python
def column_names():
    return twint.output.panda.Tweets_df.columns
def twint_to_pd(columns):
    return twint.output.panda.Tweets_df[columns]
```

```python
column_names()
tweet_df = twint_to_pd(["date", "username", "tweet", "hashtags"])
tweet_df.head(10)
```

```python
print(len(tweet_df))
```

```python
!pip install transformers
```

```python
from transformers import pipeline
sentiment_classifier = pipeline('sentiment-analysis')
```

```python
results = sentiment_classifier(tweet_df['tweet'].tolist())
```

```python
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python
len(results)
```

```python
df_sent=pd.DataFrame(results)
df_sent.head(10)
```

```python
df_combined = pd.concat([tweet_df, df_sent], axis=1, join='inner')
```

```python
df_combined.head(10)
```

```python
df_combined.to_csv("twint_dataset1.csv", index = False, encoding='utf-8')
```
