```python
df_clean = pd.read_csv('Dataset_For_Project/clean_Tweets_suicidal.csv', usecols = [1,2], encoding='latin-1')
df_clean
```

```python
df_clean.shape
```

```python
df_clean.to_csv('Dataset_For_Project/clean_Tweets_suicidal.csv', index = False, encoding='utf-8')
```

```python
df_clean.isna().any(axis=None)
```
